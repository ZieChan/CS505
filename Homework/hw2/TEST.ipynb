```python
import matplotlib.pyplot as plt

# Data from the table
num_rules_to_keep = [10, 20, 30, 40, 50, 60, 70]
BLEU_scores = [0.015644, 0.017586, 0.018972, 0.019127, 0.019321, 0.018491, 0.020154]

# Plotting the data
plt.figure(figsize=(8, 6))
plt.plot(num_rules_to_keep, BLEU_scores, marker='o', linestyle='-', color='b', label='BLEU Score')

# Adding titles and labels
plt.title('BLEU Score vs. num_rules_to_keep')
plt.xlabel('num_rules_to_keep')
plt.ylabel('BLEU Score')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
